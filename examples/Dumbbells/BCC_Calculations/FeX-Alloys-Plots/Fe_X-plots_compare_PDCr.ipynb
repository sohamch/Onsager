```python
import numpy as np
import h5py
import matplotlib.pyplot as plt
```

```python
with h5py.File("Si_data.h5","r") as fl:
    keys = fl.keys()
    Temp = np.array(fl['Temp'])
    diff_aa_Si = np.array(fl['diff_aa'])
    diff_ab_Si = np.array(fl['diff_ab'])
    diff_bb_Si = np.array(fl['diff_bb'])
    diff_bb_Si_non_local = np.array(fl['diff_bb_nl'])

with h5py.File("Cu_data.h5","r") as fl:
    keys = fl.keys()
    Temp = np.array(fl['Temp'])
    diff_aa_Cu = np.array(fl['diff_aa'])
    diff_ab_Cu = np.array(fl['diff_ab'])
    diff_bb_Cu = np.array(fl['diff_bb'])
    diff_bb_Cu_non_local = np.array(fl['diff_bb_nl'])

with h5py.File("Mn_data.h5","r") as fl:
    keys = fl.keys()
    Temp = np.array(fl['Temp'])
    diff_aa_Mn = np.array(fl['diff_aa'])
    diff_ab_Mn = np.array(fl['diff_ab'])
    diff_bb_Mn = np.array(fl['diff_bb'])
    diff_bb_Mn_non_local = np.array(fl['diff_bb_nl'])
    
with h5py.File("Ni_data.h5","r") as fl:
    keys = fl.keys()
    Temp = np.array(fl['Temp'])
    diff_aa_Ni = np.array(fl['diff_aa'])
    diff_ab_Ni = np.array(fl['diff_ab'])
    diff_bb_Ni = np.array(fl['diff_bb'])
    diff_bb_Ni_non_local = np.array(fl['diff_bb_nl'])
    
with h5py.File("Cr_data.h5","r") as fl:
    keys = fl.keys()
    Temp = np.array(fl['Temp'])
    diff_aa_Cr = np.array(fl['diff_aa'])
    diff_ab_Cr = np.array(fl['diff_ab'])
    diff_bb_Cr = np.array(fl['diff_bb'])
    diff_bb_Cr_non_local = np.array(fl['diff_bb_nl'])
```

```python
# read in Messina et. al.'s results
pdc_dat= {}
Xlist = ["Cr", "Cu", "Ni", "Si","Mn"]
for X in Xlist:
    dat=[]
    with open("PDC_ratio_%s.dat"%(X),"r") as fl:
        for line in fl:
            arr = line.split()
            dat.append(float(arr[1]))
    pdc_dat[X] = np.array(dat)
```

```python
plt.figure(figsize=(16,12))

plt.subplot(2,3,5)
plt.semilogy(1000/Temp, pdc_dat["Si"],ls='--', label="Messina et.al", linewidth=4)
plt.semilogy(1000/Temp, diff_ab_Si/diff_bb_Si_non_local, label="GF method")
# plt.semilogy(1000/Temp, diff_ab_Si/diff_bb_Si_non_local, label="GF method:%.5f THz"%(13.25951766))
plt.xlabel(r'$\mathrm{\frac{1000}{T}(K^{-1})}$', fontsize = 25)
plt.text(0.8,0.6,"X = Si", horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes,
        fontsize = 20)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize=14)

plt.subplot(2,3,3)
plt.semilogy(1000/Temp, pdc_dat["Cu"],ls='--', label="Messina et.al", linewidth=4)
plt.semilogy(1000/Temp, diff_ab_Cu/diff_bb_Cu_non_local, label="GF method",linewidth=2)
# plt.semilogy(1000/Temp, diff_ab_Cu/diff_bb_Cu_non_local, label="GF method:%.5f THz"%(1.7875284),linewidth=2)
plt.xlabel(r'$\mathrm{\frac{1000}{T}(K^{-1})}$', fontsize = 25)
# plt.ylabel(r'$\mathrm{\frac{L_{X-Fe_i}}{L_{Fe_i-Fe_i}}}$', rotation=0, labelpad = 60, fontsize = 40)
plt.text(0.8,0.6,"X = Cu", horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes,
        fontsize = 20)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize=14)

plt.subplot(2,3,2)
plt.semilogy(1000/Temp, pdc_dat["Mn"],ls='--', label="Messina et.al", linewidth=4)
plt.semilogy(1000/Temp, diff_ab_Mn/diff_bb_Mn_non_local, label="GF method",linewidth=2)
# plt.semilogy(1000/Temp, diff_ab_Mn/diff_bb_Mn_non_local, label="GF method:%.2f THz"%(2.00),linewidth=2)
plt.xlabel(r'$\mathrm{\frac{1000}{T}(K^{-1})}$', fontsize = 25)
# plt.ylabel(r'$\mathrm{\frac{L_{X-Fe_i}}{L_{Fe_i-Fe_i}}}$', rotation=0, labelpad = 60, fontsize = 40)
plt.text(0.8,0.6,"X = Mn", horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes,
        fontsize = 20)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize=14)

plt.subplot(2,3,4)
plt.semilogy(1000/Temp, pdc_dat["Ni"],ls='--', label="Messina et.al", linewidth=4)
plt.semilogy(1000/Temp, diff_ab_Ni/diff_bb_Ni_non_local, label="GF method",linewidth=2)
# plt.semilogy(1000/Temp, diff_ab_Cu/diff_bb_Cu_non_local, label="GF method:%.5f THz"%(1.7875284),linewidth=2)
plt.xlabel(r'$\mathrm{\frac{1000}{T}(K^{-1})}$', fontsize = 25)
plt.ylabel(r'$\mathrm{\frac{L_{X-Fe}}{c_xL_{Fe-Fe}}}$', rotation=0, labelpad = 60, fontsize = 40)
# plt.ylabel(r'$\mathrm{\frac{L_{X-Fe_i}}{L_{Fe_i-Fe_i}}}$', rotation=0, labelpad = 60, fontsize = 40)
plt.text(0.8,0.6,"X = Ni", horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes,
        fontsize = 20)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize=14)

plt.subplot(2,3,1)
plt.semilogy(1000/Temp, pdc_dat["Cr"],ls='--', label="Messina et.al", linewidth=4)
plt.semilogy(1000/Temp, diff_ab_Cr/diff_bb_Cr_non_local, label="GF method",linewidth=2)
# plt.semilogy(1000/Temp, diff_ab_Cu/diff_bb_Cu_non_local, label="GF method:%.5f THz"%(1.7875284),linewidth=2)
plt.xlabel(r'$\mathrm{\frac{1000}{T}(K^{-1})}$', fontsize = 25)
plt.ylabel(r'$\mathrm{\frac{L_{X-Fe}}{c_xL_{Fe-Fe}}}$', rotation=0, labelpad = 60, fontsize = 40)
# plt.ylabel(r'$\mathrm{\frac{L_{X-Fe_i}}{L_{Fe_i-Fe_i}}}$', rotation=0, labelpad = 60, fontsize = 40)
plt.text(0.8,0.6,"X = Cr", horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes,
        fontsize = 20)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize=14)

plt.tight_layout()
plt.savefig("PDC_comparison.png")
```
