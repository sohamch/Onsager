```python
import numpy as np
import h5py
import matplotlib.pyplot as plt
```

```python
with h5py.File("Si_data.h5","r") as fl:
    keys = fl.keys()
    Temp = np.array(fl['Temp'])
    diff_aa_Si = np.array(fl['diff_aa'])
    diff_ab_Si = np.array(fl['diff_ab'])
    diff_bb_Si = np.array(fl['diff_bb'])
    diff_bb_Si_non_local = np.array(fl['diff_bb_nl'])
    temp = np.array(fl['Temp'])

with h5py.File("Cu_data.h5","r") as fl:
    keys = fl.keys()
    Temp = np.array(fl['Temp'])
    diff_aa_Cu = np.array(fl['diff_aa'])
    diff_ab_Cu = np.array(fl['diff_ab'])
    diff_bb_Cu = np.array(fl['diff_bb'])
    diff_bb_Cu_non_local = np.array(fl['diff_bb_nl'])

with h5py.File("Mn_data.h5","r") as fl:
    keys = fl.keys()
    Temp = np.array(fl['Temp'])
    diff_aa_Mn = np.array(fl['diff_aa'])
    diff_ab_Mn = np.array(fl['diff_ab'])
    diff_bb_Mn = np.array(fl['diff_bb'])
    diff_bb_Mn_non_local = np.array(fl['diff_bb_nl'])
    
with h5py.File("Cr_data.h5","r") as fl:
    keys = fl.keys()
    Temp = np.array(fl['Temp'])
    diff_aa_Cr = np.array(fl['diff_aa'])
    diff_ab_Cr = np.array(fl['diff_ab'])
    diff_bb_Cr = np.array(fl['diff_bb'])
    diff_bb_Cr_non_local = np.array(fl['diff_bb_nl'])
#     diff_aa_Cr = np.array(fl['diff_aa2'])
        
with h5py.File("Ni_data.h5","r") as fl:
    keys = fl.keys()
    Temp = np.array(fl['Temp'])
    diff_aa_Ni = np.array(fl['diff_aa'])
    diff_ab_Ni = np.array(fl['diff_ab'])
    diff_bb_Ni = np.array(fl['diff_bb'])
    diff_bb_Ni_non_local = np.array(fl['diff_bb_nl'])
```

```python
Drag_dat= {}
Xlist = ["Cr", "Ni", "Cu", "Si", "Mn"]
for X in Xlist:
    dat=[]
    with open("DragRatio_%s.dat"%(X),"r") as fl:
        for line in fl:
            arr = line.split()
            dat.append(float(arr[0]))
    Drag_dat[X] = np.array(dat)
```

```python
temp = temp/1000.
```

```python
plt.figure(figsize=(16,13))

plt.subplot(2,3,1)
plt.plot(temp, Drag_dat["Cr"], ls='--', label="Messina et.al", linewidth=4)
plt.plot(temp, diff_ab_Cr/diff_aa_Cr, label="GF method")
plt.plot(temp, np.ones_like(temp)*2.417, ls='--', label="Infinite T : 2.417")
plt.xlabel(r'$T (\times 10^3 \ K)$', fontsize = 25)
plt.ylabel(r'$\mathrm{\frac{L^{XFe}}{L^{XX}}}$', rotation=0, labelpad = 60, fontsize = 40)
plt.text(0.8,0.6,"X = Cr", horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes,
        fontsize = 20)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize=14)

plt.subplot(2,3,2)
plt.plot(temp, Drag_dat["Mn"],ls='--', label="Messina et.al", linewidth=4)
plt.plot(temp, diff_ab_Mn/diff_aa_Mn, label="GF method",linewidth=2)
plt.plot(temp, np.ones_like(temp)*1.665, ls='--', label="Infinite T : 1.665")
plt.xlabel(r'$T (\times 10^3 \ K)$', fontsize = 25)
plt.text(0.8,0.6,"X = Mn", horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes,
        fontsize = 20)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize=14)

plt.subplot(2,3,3)
plt.plot(temp, Drag_dat["Cu"],ls='--', label="Messina et.al", linewidth=4)
plt.plot(temp, diff_ab_Cu/diff_aa_Cu, label="GF method",linewidth=2)
plt.plot(temp, np.ones_like(temp)*2.255, ls='--', label="Infinite T : 2.255")
plt.xlabel(r'$T (\times 10^3 \ K)$', fontsize = 25)
plt.text(0.8,0.6,"X = Cu", horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes,
        fontsize = 20)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize=14)

plt.subplot(2,3,4)
plt.plot(temp, Drag_dat["Ni"],ls='--', label="Messina et.al", linewidth=4)
plt.plot(temp, diff_ab_Ni/diff_aa_Ni, label="GF method")
plt.plot(temp, np.ones_like(temp)*2.238, ls='--', label="Infinite T : 2.238")
plt.xlabel(r'$T (\times 10^3 \ K)$', fontsize = 25)
plt.ylabel(r'$\mathrm{\frac{L^{XFe}}{L^{XX}}}$', rotation=0, labelpad = 60, fontsize = 40)
plt.text(0.8,0.6,"X = Ni", horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes,
        fontsize = 20)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize=14)

plt.subplot(2,3,5)
plt.plot(temp, Drag_dat["Si"],ls='--', label="Messina et.al", linewidth=4)
plt.plot(temp, diff_ab_Si/diff_aa_Si, label="GF method")
plt.plot(temp, np.ones_like(temp)*1.482, ls='--', label="Infinite T : 1.482")
plt.xlabel(r'$T (\times 10^3 \ K)$', fontsize = 25)
plt.text(0.8,0.6,"X = Si", horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes,
        fontsize = 20)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize=14)

plt.tight_layout()
plt.savefig("FCR_comparison.png")
```
